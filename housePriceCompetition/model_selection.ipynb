```python
def miles_converter(mileage):
    mileage = mileage.replace(',', '').strip()
    if '(mile)'in mileage:
        mileage = mileage.replace('(mile)', '')
        mileage =  int(np.round(int(mileage) / 0.621371))
    return mileage
```

```python
df['Mileage(km)'] = df['Mileage(km)'].apply(miles_converter)
```

```python
def remove_save(price):
    if 'Save' in price:
        index = price.find('S')
        price = price[0:index]
    price = price.replace(',', '')
    return price
```

```python
df['Price($)'] = df['Price($)'].apply(remove_save)
```

```python
df = df[df['Steering'].isin(['Right','Left'])]
df = df[df['Transmission'].isin(['AT','MT'])]
```

```python
df = df.replace('-', np.NaN)
df = df.replace(' ', np.NaN)
df = df.replace('', np.NaN)
df = df.dropna()
```

```python
X_cols = df.drop('Price($)', axis=1)
y_cols = df['Price($)']
```

```python
from sklearn.svm import SVC
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
```

```python
models = {

    'linear Reg': {
        'model' : LinearRegression(),
        'params': {}
    },
    'knn': {
        'model': KNeighborsClassifier(),
        'params': {
            'n_neighbors': [1, 20, 30, 15],
            'metric': ['euclidean', 'minkowski']
        }
    },
    'random forest':{
        'model': RandomForestClassifier(),
        'params': {
            'n_estimators' : [2, 7, 20, 15, 10]
        }
    },
    'decision tree':{
        'model': DecisionTreeClassifier(class_weight='balanced'),
        'params': {
            'criterion' :['gini', 'entropy']
        }
    }
}
```

```python
from sklearn.model_selection import StratifiedKFold

sf = StratifiedKFold(n_splits=5)
```

```python
results = []
for model, model_params in models.items():
    my_model = GridSearchCV(model_params['model'], model_params['params'], cv=sf)
    my_model.fit(X_cols, y_cols)
    results.append({
        'model name': model,
        'best score': my_model.best_score_,
        'best params': my_model.best_params_
    })
    
    
```

```python
results_df = pd.DataFrame(results)
results_df
```
