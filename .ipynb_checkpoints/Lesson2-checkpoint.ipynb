# Mutual Information

## Where Do I Even Begin?

We all have that moment when we are given a huge data set with tens of columns and variables to work with. We wonder how to even begin from where to begin which features to use, which model to use, whats the relation between the features and the target????

## A Good Place to Start

The best place to start is, ranking the features and selecting the ones with a greater the best or hightest relationship to the target variable. This will reduce the total number of columns and features you have to work with by selecting the very best feature that have a high correlation to the target.

## Mutual Information

This is a metric that will help us understand the correlation between two quantities. Unlike the ordinary correlation that only explains the linear relationship, mutual information explains the non linear relationship as well. Mutual information can detect any kind of relationship.

When you are given a data set and little to no description is provide, the best metric to use and a good starting place is the mutual information metric that lets you in on the relationships and features that relate best to the target.

## Why you would use mutual Information

- theoretically well-founded
- easy to use and interpret
- resistant to overfitting
- able to detect any kind of relationship
- computationally efficient

## What is mutual Information (MI) 

The mutual information between two quantities is a measure of the extent to which knowledge of one quantity reduces uncertainty about the other.

It answers the question of, **If you knew the value of a feature, how much more confident would you be about the target?**

This basically means knowing a certain feature will reduce uncertainty about another feature. Uncertainty is basically the **entropy**.

**Entropy** is the degree of randomness of a dataset.

## Interpreting Mutual Information

The lowe bound of mutual information is 0 while there is not limit to the upper bound.

With 0 means the is not relationship between the variables, they are independent of each other and knowing one variable, you can not tell or answer any question about the other variable. Any value above one means there is a relationship between the two variables.

- MI can help you to understand the relative potential of a feature as a predictor of the target, considered by itself.

- It's possible for a feature to be very informative when interacting with other features, but not so informative all alone. MI can't detect interactions between features. It is a univariate metric.

- The actual usefulness of a feature depends on the model you use it with. A feature is only useful to the extent that its relationship with the target is one your model can learn. Just because a feature has a high MI score doesn't mean your model will be able to do anything with that information. You may need to transform the feature first to expose the association.

[Source](https://www.kaggle.com/ryanholbrook/mutual-information)

## We'll use an automobile Data set To Study Mutual Information

We'll be using sklearn, and it comes with two mutual information classes from the feature_selection module, one for real-valued targets(mutual_info_regression) and the other one for categorical valued targets(mutual_info_classif)

scikit learn treats MI as discret or continuous. Hence we need to specify explicitly whether its discret or continuous. We can do generalily consider float dtype as continuous and categorical types as discret after performing label encoding.

## Code section

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("datasets/Automobile_data.csv")
```

```python
df.head()
```

```python
len(df.columns)
```

```python
X = df.copy()
y = X.pop("price")

# Label encoding for categoricals
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.factorize.html
for colname in X.select_dtypes("object"):
    X[colname], _ = X[colname].factorize()

# All discrete features should now have integer dtypes (double-check this before using MI!)
discrete_features = X.dtypes == int
```

```python
y = [val for val in y if val != '?']
y
X = X[:len(y)]
```

```python
X
```

```python
from sklearn.feature_selection import mutual_info_regression

def generate_mi_score(X, y, discrete_features):
    mi_score = mutual_info_regression(X, y, discrete_features = discrete_features)
    mi_score = pd.Series(mi_score, name = "mi_scores", index = X.columns)
    mi_score = mi_score.sort_values(ascending=False)
    return mi_score
```

```python
mi_score = generate_mi_score(X, y, discrete_features)
```

```python
mi_score
```

```python
plt.style.use('seaborn-whitegrid')

def plot_utility_score(scores):
    y = scores.sort_values(ascending=True)
    width = np.arange(len(y))
    ticks_labels = list(y.index)
#   https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.barh.html
# plt.bar(how_many_elements, width_size_of_bars)
    plt.barh(width, y)
    plt.yticks(width, ticks_labels)
    plt.title("Mutual Information Scores")

# dpi >> dot per inch, determines how many pixels the figure comprises. 
plt.figure(dpi=100, figsize=(8, 5))
plot_utility_score(mi_score)
```
