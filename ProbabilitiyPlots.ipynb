## 1. Quantile Quantile Plots(Q-Q Plots) Or Normal Probability Plots

A Q-Q plot helps to determine the shape of a distribution. Example if you wish to know if your dataset is a normal distribution, you can use a Q-Q plot to help answer these question.

[Video explanation StatQuest](https://www.youtube.com/watch?v=okjYjClSjOg)

[Video explanation 2](https://www.youtube.com/watch?v=smJBsZ4YQZw)

[Wikipedia](https://en.wikipedia.org/wiki/Probability_plot)

```python
from scipy import stats
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
data = np.random.normal(loc = 1, scale = 3, size = 1000)
```

```python
sns.distplot(data)
```

```python
res = stats.probplot(data, plot = plt)
```

The more the dots are closer to the line the more the data is normaly distributed and vice versa.
